# Scaling

```python
# scaling >> refers to feature scaling in which we format the features to a specific range so that no feature dominates others due to its scale.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

import warnings 
warnings.filterwarnings("ignore")
```

```python
# feature scaling
#standardisation >> mu=0, sigma=1
#normalisation(minmax scaling) >> data comes between 0 and 1
#unit vector>> 1
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.distplot(df["total_bill"])
```

```python
mean = np.mean(df['total_bill'])
std = np.std(df["total_bill"])
```

```python
std_data = []
for i in list(df['total_bill']):
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
print(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']])
```

```python
print(scaler.transform(df[['total_bill']]))
```

```python
df
```

```python
scaler.fit_transform(df[['total_bill', "tip"]])
```

```python
df2 = pd.DataFrame(scaler.fit_transform(df[['total_bill', "tip"]]), columns = ['total_bill', 'tip'])
```

```python
df2
```

```python
# transform is used both on trained and test data wheras fit is used on;ly ontrain data


scaler.transform([[13,12]]) # test data
```

```python
# normalization  /min/max scaling 

from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[["total_bill", 'tip']])
```

```python
min_max.transform([[10,2]])
```

```python
# 3. unit vector

from sklearn.preprocessing import normalize
```

```python
uv = normalize(df[["total_bill"]])
uv # since one dimension so all value are 1
```
