```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_regression

X,y = make_regression(n_samples = 1000 , n_features = 2,noise = 10, random_state = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor

reg = KNeighborsRegressor()
reg.fit(X_train , y_train)
```

```python
y_pred =  reg.predict(X_test)
y_pred
```

```python
# Evaluate the model
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("Current model performance:")
print(f"R2 score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
```

```python
# hyperparameter tuning 
from sklearn.model_selection import GridSearchCV
param_grid = {
    "n_neighbors":[3,5,6,7,9,11,13],
    "weights":["uniform" , "distance"],
    "algorithm":["auto", "ball_tree", "kd_tree", "brute"],
    "leaf_size":[20,30,40,50]
}
```

```python
grid = GridSearchCV(estimator = reg , param_grid = param_grid , cv = 5 ,verbose = 3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
# predict with best estimator
best_model = grid.best_estimator_
y_pred_tuned = best_model.predict(X_test)
y_pred_tuned
```

```python
# Evaluaate the model 
print("current model Performance ")
print(f"R2_score : {r2_score(y_test, y_pred_tuned)}")
print("mean_absolute_error :")
print(mean_absolute_error(y_test, y_pred_tuned))
print("mean_squared_error :")
print(mean_squared_error(y_test , y_pred_tuned) )
```
