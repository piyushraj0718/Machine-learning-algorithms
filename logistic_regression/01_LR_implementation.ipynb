```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
# logistic regression till now , we have studied , with two values of the target variable>> binary classification
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df["target"] = data.target
df
```

```python
df.target.unique()
```

```python
# for binary classification >> only two class neede, so we will filter out class 2 

df = df[df["target"] != 2]
```

```python
df.target.value_counts()
```

```python
# split the dataa into x and y 

x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# train_test_split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# scaling 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# model training
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
classifier.predict_proba(x_test) # corresponding to higher probability class is predicted , cutoff is 0.5
```

```python
# evaluation metrics 

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print( classification_report(y_test, y_pred) )
```

```python
#first two rows, 0 and 1 >> precision recall, f1 score and no of data points with respect to respective class
#macroavg >> for each class take simple avg>> avg precision of class 0 and class 1 will be macro average
#similarly for recall, f1 score
#support will be total no of datapoints

#weighted avg is used in case of class imbalance, it gives importance to performance of dominant class
```

```python
from sklearn.metrics import roc_curve , auc

# get predicted probabilities of class1 
 
Y_pred_proba = classifier.predict_proba(x_test)[:,1]  # probability of one class

Y_pred_proba
```

```python
#reciever operating characteristics curve
#inputs : y_test, y_pred_proba. It will return TPR, FPR with different cutoff of probabilities


fpr, tpr, thresholds = roc_curve(y_test , Y_pred_proba)
```

```python
fpr
```

```python
tpr
```

```python
thresholds
```

```python
# area under curve
roc_auc =  auc(fpr, tpr)
roc_auc
```

```python
# plot roc_auc curve 

plt.figure(figsize = (8, 6)) # create a new fig with size 8,6
plt.plot(fpr, tpr, color='orange', label='ROC curve (area = %0.2f)' % roc_auc) # plot fpr vs tpr with label
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--') # dashed diagonal line
plt.xlim([0, 1]) # x axis limits
plt.ylim([0, 1.05]) # y axis limits
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc= "lower right")
plt.show()
```
