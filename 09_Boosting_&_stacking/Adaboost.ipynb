# Adaboost Classifier and Regressor

```python
# Adabosst classifier

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split


from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


import warnings 
warnings.filterwarnings("ignore")
```

```python
X , y = make_classification(n_samples = 1000 , n_features = 20 , n_classes = 2 , random_state=42) 
```

```python
X
```

```python
X_train ,X_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=42)
```

```python
classifier = AdaBoostClassifier()
# estimator is the number of weak learners we want to use in our model.
classifier.fit(X_train , y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# evaluate the model
print("Current model Performance :")
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0,1.5,2.0]# samme is the original adaboost algorithm and SAMME.R is a variant that uses a different approach to calculate the weights of the weak learners.
}
```

```python
ada = AdaBoostClassifier()
grid_search = GridSearchCV(estimator=ada, param_grid=param_grid, cv=5, verbose=3) 
grid_search
```

```python
grid_search.fit(X_train , y_train)
```

```python
grid_search.best_params_
```

```python
bestmodel = grid_search.best_estimator_
```

```python
y_model_pred = bestmodel.predict(X_test)
y_model_pred
```

```python
# evaluate the model
print("Current model Performance :")
accuracy = accuracy_score(y_test, y_model_pred)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, y_model_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_model_pred))
```

```python
# adaboost regression 

from sklearn.datasets import make_regression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score , mean_absolute_error


X , y = make_regression(n_samples = 1000 , n_features = 2, noise = 10, random_state=42)
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
reg = AdaBoostRegressor()
reg.fit(X_train , y_train)
```

```python
y_regpred = reg.predict(X_test)

# evaluate the model
print("Current model Performance :")
print(f"R2 score :{r2_score(y_test, y_regpred)}")
print(f"Mean Squared Error :{mean_squared_error(y_test, y_regpred)}")   
print(f"and Mean Absolute Error :{mean_absolute_error(y_test, y_regpred)}")
```

```python
# hypper parameter tuning for regression

from sklearn.model_selection import GridSearchCV
param_grid= {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0,1.5,2.0],
    'loss': ['linear', 'square', 'exponential']} #    loss function to use when updating the weights after each iteration. The options are linear, square, and exponential.}

grid_search = GridSearchCV(estimator=reg, param_grid=param_grid, cv=5, verbose=3)
grid_search.fit(X_train , y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)
```

```python
# Evaluate the tuned model
print("Tuned model performance:")
print(f"R2 score: {r2_score(y_test, y_pred_tuned)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred_tuned)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred_tuned)}")
```
