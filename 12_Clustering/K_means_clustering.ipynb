```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples = 1000 , centers=3, n_features=2,random_state=1)
```

```python
X
```

```python
# in usl you will not use y  
```

```python
plt.scatter(X[:,0], X[:,1] ,color = "yellow")
```

```python
plt.scatter(X[:,0], X[:,1] ,c = y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train ,y_test = train_test_split(X,y , test_size= 0.30 , random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.cluster import KMeans

KMeans= KMeans(n_clusters=3)
KMeans
```

```python
KMeans.fit(X_train)
```

```python
y_labels = KMeans.predict(X_train)
y_labels
```

```python
plt.scatter(X_train[:,0], X_train[:,1] ,c = y_labels)
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_train , KMeans.labels_)
```

```python
# WCSS distance  (k=3)

KMeans.inertia_
```

```python
# how to select k (centroid) >> elbow method
from sklearn.cluster import KMeans
wcss = []
for k in range(1,10):
    Kmeans = KMeans(n_clusters = k)
    Kmeans.fit(X_train)
    wcss.append(Kmeans.inertia_)
```

```python
wcss
```

```python
# plot the elbow curve 

plt.plot(range(1,10), wcss)
plt.xticks(range(1,10))
plt.xlabel("no. of students")
plt.ylabel("wcss")
plt.show()
```

```python
# based on the plot above the optimal no. of k/centroid should be 3
```
