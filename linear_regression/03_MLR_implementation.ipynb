```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
dia = load_diabetes()
dia
```

```python
print(dia.DESCR)
```

```python
data = pd.DataFrame(dia.data , columns = dia.feature_names)
data
```

```python
data["target"] = dia.target
data
```

```python
# EdA and data preparation
```

```python
x = data.drop("target", axis = 1)
y = data['target']
```

```python
x
```

```python
y
```

```python
# split the data into train and test

from sklearn.model_selection import train_test_split
x_train , x_test , y_train, y_test = train_test_split(x , y , test_size = 0.2 , random_state = 1)
```

```python
x_train.shape , x_test.shape
```

```python
# scaling

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler
```

```python
# scaling should be done always after train test split
# target varible should not  be scaled 
# avoid scaling categorincal features 


x_train = scaler.fit_transform(x_train) # for train data use fit_transform
x_test = scaler.transform(x_test)             # for test data use transform
```

```python
x_train
```

```python
x_test
```

```python
# model training 

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train , y_train)
```

```python
x_train.shape
```

```python
model.coef_
```

```python
len(model.coef_)
```

```python
model.intercept_
```

```python
#interpretation
#say age has coef value -30.62219046
#interpretation> 1 unit change in age then target value decreases by -30 units on an average when all
#other features/factors are kept constant
x_train
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
