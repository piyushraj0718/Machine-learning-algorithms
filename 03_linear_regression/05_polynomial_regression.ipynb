```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate some synthetic dp  or reproducibility of results 
np.random.seed(1)
```

```python
# generate synthetic X
x= 2*np.random.rand(100,1) # 100 samples, single feature
```

```python
x
```

```python
# generate random synthetic y using x

y = 4+3*x +1.5*x**2 + np.random.randn(100,1)
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=1)
```

```python
x_train.shape , x_test.shape, y_train.shape, y_test.shape
```

```python
#  perform polynomial regression

from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2 , include_bias= False)
x_poly_train = poly.fit_transform(x_train)
```

```python
x_poly_train
```

```python
from sklearn.linear_model import LinearRegression
poly_reg = LinearRegression()
poly_reg
```

```python
poly_reg.fit(x_poly_train,y_train)
```

```python
poly_reg.coef_
```

```python
poly_reg.intercept_
```

```python
# make prediction on train data itself for the evaluation metrics

y_polypred = poly_reg.predict(x_poly_train)
y_polypred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, y_polypred)
```

```python
plt.scatter(x_train,y_train)
```

```python
degree = 2
plt.scatter(x_train, y_train, label = "training data")
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
plt.plot(X_range, poly_reg.predict(X_range_poly), color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
